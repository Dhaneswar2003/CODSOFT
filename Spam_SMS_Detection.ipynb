```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_data = pd.read_csv("/content/sample_data/spam.csv", encoding='latin-1')
```

```python
print(raw_data)
```

```python
raw_data.shape
```

```python
sms_data  = raw_data.where((pd.notnull(raw_data)),'')
```

```python
sms_data.head()
```

```python
sms_data.rename({'v1':'Category','v2':'Message'},inplace=True,axis=1)
sms_data.head()
```

```python
sms_data.describe()
```

```python
sms_data.duplicated().sum()
```

```python
df = sms_data.drop_duplicates(keep= 'first').copy()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df.loc[df['Category'] == 'spam' , 'Category',] = 0
df.loc[df['Category'] == 'ham' , 'Category',] = 1
```

```python
df.head()
```

```python
X = df['Message']
Y = df['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train ,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features , Y_train)
```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train,prediction_on_training_data)
```

```python
print("Accuracy on Training Data: ",accuracy_on_training_data)
```

```python
input_SMS = ["TheMob> Check out our newest selection of content, Games, Tones, Gossip, babes and sport, Keep your mobile fit and funky text WAP to 82468,"]
```

```python
input_data_features = feature_extraction.transform(input_SMS)
```
